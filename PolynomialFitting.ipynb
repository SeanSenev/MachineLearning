```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
plt.style.use('seaborn-whitegrid')
import numpy as np

noise_scale = 100
number_of_samples = 50
```

```python
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
```

```python
x_values = x.flatten()
y_values = y.flatten()
```

Part 1

```python
# N-th order polynomials
m1 = np.poly1d(np.polyfit(x_values, y_values, 1))
m2 = np.poly1d(np.polyfit(x_values, y_values, 2))
m3 = np.poly1d(np.polyfit(x_values, y_values, 3))
m4 = np.poly1d(np.polyfit(x_values, y_values, 4))
m5 = np.poly1d(np.polyfit(x_values, y_values, 5))
m6 = np.poly1d(np.polyfit(x_values, y_values, 6))
m7 = np.poly1d(np.polyfit(x_values, y_values, 7))
m8 = np.poly1d(np.polyfit(x_values, y_values, 8))
```

Plotting polynomial functions over a set of data points to see which fits the data the best 

```python
plt.figure(figsize=(15,9))
plt.plot(x, y, 'ro')
line = np.linspace(-20, 5, 100)
plt.plot(line, m1(line))
plt.plot(line, m2(line))
plt.plot(line, m3(line))
plt.plot(line, m4(line))
plt.plot(line, m5(line))
plt.plot(line, m6(line))
plt.plot(line, m7(line))
plt.plot(line, m8(line))
```

```python
# calcuating mean squared error
mse1 = mean_squared_error(y_values, m1(x_values))
mse2 = mean_squared_error(y_values, m2(x_values))
mse3 = mean_squared_error(y_values, m3(x_values))
mse4 = mean_squared_error(y_values, m4(x_values))
mse5 = mean_squared_error(y_values, m5(x_values))
mse6 = mean_squared_error(y_values, m6(x_values))
mse7 = mean_squared_error(y_values, m7(x_values))
mse8 = mean_squared_error(y_values, m8(x_values))

overall_mse = [mse1, mse2, mse3, mse4, mse5, mse6, mse7, mse8]
order = [1, 2, 3, 4, 5, 6, 7, 8]
```

Plot of MSE vs Polynomial Order

```python
plt.figure(figsize=(15,9))
plt.ylim(0, 80000)
plt.plot(order, MSE)
plt.title('Mean Squared Error vs Order')
plt.xlabel('Polynomial Degree')
plt.ylabel('Error')
```
